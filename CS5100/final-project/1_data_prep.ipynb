```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

cols = [
    "age", "sex", "cp", "trestbps", "chol", "fbs", "restecg",
    "thalach", "exang", "oldpeak", "slope", "ca", "thal", "num",
]

# Read raw UCI file and treat '?' as missing
df = pd.read_csv("processed.cleveland.data", names=cols, na_values="?")

# Drop rows with missing ca / thal (only a few rows)
df_clean = df.dropna().copy()

# Binary target: 0 = no disease, 1 = any disease (num 1–4)
df_clean["target"] = (df_clean["num"] > 0).astype(int)

print(df_clean["target"].value_counts(normalize=True))

sns.countplot(x="target", data=df_clean)
plt.show()

df_clean.to_csv("cleveland_cleaned.csv", index=False)
```
